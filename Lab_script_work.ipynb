```python
import numpy as np
import matplotlib.pyplot as plt

p=np.linspace(1,100,100)
n=1.001
m1=0.14
m2=0.49
theta1=np.arccos(np.sqrt(m1**2+p**2)/(n*p))
theta2=np.arccos(np.sqrt(m2**2+p**2)/(n*p))
no_theta_2=np.argwhere(~np.isnan(theta2))
no_theta_1=np.argwhere(~np.isnan(theta1))
fig,ax=plt.subplots(1,figsize=(8,6))
ax.plot(p,theta1,label='pion')
ax.plot(p,theta2,label='kaon')
if no_theta_1[0]>no_theta_2[0]:
    ax.axvspan(no_theta_1[0][0]+1,np.argwhere(abs(theta2-theta1)<0.001)[0][0],color='grey',alpha=0.2,label='distinguishing region')
else:
    ax.axvspan(no_theta_2[0][0]+1,np.argwhere(abs(theta2-theta1)<0.001)[0][0],color='grey',alpha=0.2,label='distinguishing region')
ax.set_xlabel('momentum')
ax.set_ylabel('theta')
fig.suptitle('Cherenkov angle against momentum')
ax.legend()
plt.savefig('Cherenkov')
```
